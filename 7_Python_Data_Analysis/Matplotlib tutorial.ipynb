```python
# Create a new figure of size 8x6 points, using 100 dots per inch ...figsize=(width, height)
#出力されるグラフ画像の大きさ
plt.figure(figsize=(8,6), dpi=80)

# Create a new subplot from a grid of 1x1 異なるグラフを、別々に描画し、１つの図にしたい場合
plt.subplot(1,1,1) #plt.subplot(111) plt.subplot(行数, 列数, 何番目のプロットか)


X = np.linspace(-np.pi, np.pi, 256,endpoint=True)
C,S = np.cos(X), np.sin(X)

# Plot cosine using blue color with a continuous line of width 1 (pixels)
plt.plot(X, C, color="orange", linewidth=1.0, linestyle="-")

# Plot sine using green color with a continuous line of width 1 (pixels)
plt.plot(X, S, color="green", linewidth=1.0, linestyle="-")

# Set x limits ...x軸の両端
plt.xlim(-6.0,6.0)

# Set x ticks　目盛りの設定
plt.xticks(np.linspace(-6,6,13,endpoint=True))

# Set y limits ...x軸の両端
plt.ylim(-1.0,1.0)

# Set y ticks 目盛りの設定
plt.yticks(np.linspace(-1,1,5,endpoint=True))

# Save figure using 72 dots per inch
# savefig("../figures/exercice_2.png",dpi=72)

# Show result on screen
plt.show()
```

```python
# plt.subplot(行数, 列数, 何番目のプロットか)
# sinのプロット
# plt.subplot(1, 2, 1)
plt.subplot(2, 1, 1)
plt.plot(X, S)
plt.title(r"$\sin x$")
plt.xticks(
    np.linspace(-np.pi, np.pi,5,endpoint=True),
    [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$']
          )
# ↑$ドルマークで囲まれた範囲はtex形式になる$

plt.yticks(
    [-1,0,1],
    [r'$minus\ one$', r'$zero$', r'$plus\ one$']
          )
# ↑as tex notation, put \ before space to output the space

plt.xlim(-np.pi, np.pi)
plt.ylim(-1.3, 1.3)

# cosのプロット
# plt.subplot(1, 2, 2)
plt.subplot(2, 1, 2)
plt.plot(X, C)
plt.title(r"$\cos x$")
plt.xlim(-np.pi, np.pi)
plt.ylim(-1.3, 1.3)

plt.tight_layout()  # タイトルの被りを防ぐ
```

```python
plt.figure(figsize=(8,6), dpi=80)
plt.subplot(1,1,1) #plt.subplot(111) plt.subplot(行数, 列数, 何番目のプロットか)
X = np.linspace(-np.pi, np.pi, 256,endpoint=True)
C,S = np.cos(X), np.sin(X)
plt.plot(X, C, color="orange", linewidth=1.0, linestyle="-")
plt.plot(X, S, color="green", linewidth=1.0, linestyle="-")

plt.xlim(-6.0,6.0)
plt.xticks(np.linspace(-6,6,13,endpoint=True))
plt.ylim(-1.0,1.0)
plt.yticks(np.linspace(-1,1,5,endpoint=True))

# spine は axis の目盛のマーカーをつなげる線で、データ領域の境界を示す
# デフォルトでは axis の境界に位置していた
ax = plt.gca()
# 上と右の色無くして非表示にする
ax.spines['right'].set_color('none') #.set_color('none')で非表示にできる
ax.spines['top'].set_color('none')
# 　下と左をデータの空間座標の 0 にします。
ax.xaxis.set_ticks_position('bottom') # 下の軸をx軸として使う
ax.spines['bottom'].set_position(('data',0)) #下の軸の位置を上下中央に持ってくる
# ax.spines['bottom'].set_position(('center'))
ax.yaxis.set_ticks_position('left') # 左の軸をy軸として使う
ax.spines['left'].set_position(('data',0))#左の軸の位置を左右中央に持ってくる

plt.show()
```

<h2>set_position(position)</h2>
set the position of the spine

Spine position is specified by a 2 tuple of (position type, amount). The position types are:

<ul>
<li>‘outward’ : place the spine out from the data area by the specified number of points. (Negative values specify placing the spine inward.)</li>
<li>‘axes’ : place the spine at the specified Axes coordinate (from 0.0-1.0).</li>
<li>‘data’ : place the spine at the specified data coordinate.</li>
</ul>
Additionally, shorthand notations define a special positions:
‘center’ -> (‘axes’,0.5)
‘zero’ -> (‘data’, 0.0)

```python
plt.figure(figsize=(8,6), dpi=80)

plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cosine")
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="-", label="sine")
plt.legend(loc='upper left', frameon=False)
ax = plt.gca()
ax.spines['right'].set_color('none') #.set_color('none')で非表示にできる
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom') # 下の軸をx軸として使う
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left') 
ax.spines['left'].set_position(('data',0))
#-----
t = 2*np.pi/3
plt.plot([t,t],[0,np.cos(t)], color ='blue', linewidth=1.5, linestyle="--") #dotted line
plt.scatter([t],[np.cos(t)], 60, color ='blue') # point on the curve
#scatterの第三引数はponts単位でのマーカーのサイズ

plt.annotate(r'$\sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
             xy=(t, np.sin(t)), xycoords='data',
             xytext=(+10, +30), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.plot([t,t],[0,np.sin(t)], color ='red', linewidth=1.5, linestyle="--")
plt.scatter([t,],[np.sin(t),], 50, color ='red')

plt.annotate(r'$\cos(\frac{2\pi}{3})=-\frac{1}{2}$',
             xy=(t, np.cos(t)), xycoords='data',
             xytext=(-90, -50), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.65 ))
    
plt.show()
```

```python
n = 256
X = np.linspace(-np.pi,np.pi,n,endpoint=True)
Y = np.sin(2*X)

# plt.axes([0.025,0.025,0.95,0.95])

plt.plot (X, Y+1, color='blue', alpha=1.00)
plt.fill_between(X, 1, Y+1, color='blue', alpha=.25)

plt.plot (X, Y-1, color='blue', alpha=1.00)
plt.fill_between(X, -1, Y-1, (Y-1) > -1, color='blue', alpha=.25)
plt.fill_between(X, -1, Y-1, (Y-1) < -1, color='red',  alpha=.25)

plt.xlim(-np.pi,np.pi), plt.xticks([])
plt.ylim(-2.5,2.5), plt.yticks([])
# savefig('../figures/plot_ex.png',dpi=48)
plt.show()
```

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
T = np.arctan2(Y,X)
#arctan2(x,y)= (x,y)座標の角度θ

plt.axes([0.025,0.025,0.95,0.95])
plt.scatter(X,Y, s=75, c=T, alpha=.5)
print(T[:10])
plt.xlim(-1.5,1.5), plt.xticks(np.linspace(-1.5,1.5,9,endpoint=True))
plt.ylim(-1.5,1.5), plt.yticks([])
# savefig('../figures/scatter_ex.png',dpi=48)
plt.show()
```

```python

n = 12
X = np.arange(n)
Y1 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)
Y2 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)

plt.axes([0.025,0.025,0.95,0.95])
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

for x,y in zip(X,Y1):
    plt.text(x+0.4, y+0.05, '%.2f' % y, ha='center', va= 'bottom')

for x,y in zip(X,Y2):
    plt.text(x+0.4, -y-0.05, '%.2f' % y, ha='center', va= 'top')

plt.xlim(-.5,n), plt.xticks([-.5,(n-0.5)/2,n])
plt.ylim(-1.25,+1.25), plt.yticks([])

# savefig('../figures/bar_ex.png', dpi=48)
plt.show()
```

```python
import matplotlib.pyplot as plt

ax = plt.axes([0.025,0.025,0.95,0.95])

ax.set_xlim(0,4)
ax.set_ylim(0,3)
ax.xaxis.set_major_locator(plt.MultipleLocator(1.0))
ax.xaxis.set_minor_locator(plt.MultipleLocator(0.1))
ax.yaxis.set_major_locator(plt.MultipleLocator(1.0))
ax.yaxis.set_minor_locator(plt.MultipleLocator(0.1))
ax.grid(which='major', axis='x', linewidth=0.75, linestyle='-', color='0.75')
ax.grid(which='minor', axis='x', linewidth=0.25, linestyle='-', color='0.75')
ax.grid(which='major', axis='y', linewidth=0.75, linestyle='-', color='0.75')
ax.grid(which='minor', axis='y', linewidth=0.25, linestyle='-', color='0.75')
ax.set_xticklabels([])
ax.set_yticklabels([])

# savefig('../figures/grid_ex.png',dpi=48)
plt.show()
```
