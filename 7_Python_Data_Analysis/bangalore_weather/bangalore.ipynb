```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('bangalore_temparature.tsv',sep='\t')
print(df.shape)
print(df.describe())
```

```python
df.hist(column='Mar')

def compare_hist(month1,month2):
    f, (ax1) = plt.subplots(1)
    df.hist(column=month1, bins=10, ax=ax1)
    df.hist(column=month2, bins=10, ax=ax1, color='red')
    plt.xlabel('Temp')
    plt.ylabel('# of Years')
    plt.title(month1 +' vs '+month2)

# print('ssss')
# print(plt.subplots(1))    
compare_hist('Mar','Apr')
```
