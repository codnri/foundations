```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
df = pd.read_csv("recent-grads.csv")
print(df.head())
print(df.tail())
clean_df= df.dropna()
clean_df
```

## Assignment

- Create a for loop that plots the histogram of sample size, median, employed, full time, unemployment rate, men and women (remember to use the exact column titles)

```python
cols = ['Sample_size', 'Men','Women', 'Employed', 'Full_time', 'Unemployment_rate', 'Median']

fig = plt.figure(figsize=(7,12))
for i in range(len(cols)):
    ax = fig.add_subplot(len(cols),2,i+1)
    ax.set_xlabel(cols[i])
    ax = clean_df[cols[i]].plot(kind='hist')

fig.subplots_adjust(wspace=0.4, hspace=0.6)
# for col in cols:
#     clean_df.hist(column=col,figsize=(8,8),color="blue")
 
    
```
